Decision tree with iris data
Post prunning

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
import seaborn as sns

iris = load_iris()

# print(iris)
# print(iris.data)  #only input data

# output features 
display(iris.target)

#seaborn
df = sns.load_dataset('iris')
df.head()

# @independent feature and dependent features 
x = df.iloc[:,:-1]
y = iris.target 

print('\n -----independent features----------------\n',x)
print('\n ----------dependent features------------\n',y)
```

```python
#---train test split
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.33, random_state=42)
# x_train

#---post pruning decision
#decision tree model
treemodel = DecisionTreeClassifier()

#--train the model
treemodel.fit(x_train, y_train)

# DecisionTreeClassifier()


#constract the tree
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)
```

```python
#---Used---max_depth
treemodel = DecisionTreeClassifier(max_depth=2)

#--train the model
treemodel.fit(x_train, y_train)

# DecisionTreeClassifier()


#constract the tree
from sklearn import tree
plt.figure(figsize=(15, 10))
tree.plot_tree(treemodel, filled=True)

#prediction
y_pred = treemodel.predict(x_test)
print(y_pred)

#check accuracy
from sklearn.metrics import accuracy_score, classification_report

score = accuracy_score(y_pred, y_test)
print("\n Accuracy score is :",score)

#classification_report
print("\n Classification report :",classification_report(y_pred, y_test))
```

PRE PRUNNING IN DECISION TREE

```python
#Preprunning
parameter ={
    'criterion' :['gini', 'entropy', 'log_loss'],
    "splitter" :['best', 'random'],
    "max_depth":[1, 2, 3, 4, 5],
    "max_features":['auto','sqrt', 'log2'],
    # "ccp_alpha":[1, 2, 3, 4, 5, 6, 7]
}

#GridSearchCv
from sklearn.model_selection import GridSearchCV

treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel, param_grid=parameter, cv=5, scoring='accuracy')

cv.fit(x_train, y_train)

print(cv.best_params_)


#prediction
y_pred = cv.predict(x_test)
print("\n predict is :\n",y_pred)

#score
print("\n accuracy_score : ",accuracy_score(y_pred, y_test))

#classification report
print("\n classification_report : \n",classification_report(y_pred, y_test))


```
