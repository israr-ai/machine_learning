```python
#Linear regreassion
#Example: “Hours studied vs Marks scored”
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import  LinearRegression

# step-1  Training data (Hours and Marks)
x_train = np.array([[1], [2], [3], [4], [5], [6], [7]]) # Hours studied = Input feature
y_train = np.array([10, 20, 30, 40, 50, 60, 70]) # Marks scored Target variable (output label)

# step-2 Train linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# step-3 predict for new input
x_test = np.array([[5]]) #new data ->6 hours studied
y_pred = model.predict(x_test)

print(f"Prediction for 6 hours study : {y_pred[0]} marks")


# ------------------------------
# Step 4: Plot Training Data + Regression Line
# ------------------------------
# plt.scatter(x_train, y_train, color='blue', label='Training Data (Actual)')

plt.scatter(x_train, y_train, color='blue', label='Training Data (Actual)')
plt.plot(x_train, model.predict(x_train), color='red', label='Regression Line')
plt.scatter(x_test, y_pred, color='green', s=100, label=f'Predicted Point (6 hrs = {y_pred[0]:.1f})')

plt.title('Linear Regression Example: Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks Scored')
plt.legend()
plt.grid(True)
plt.show()

```

```python
#--Logistic regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Step 1: Data
X = np.array([[1], [2], [3], [4], [5], [6]])  # Hours studied
y = np.array([0, 0, 0, 1, 1, 1])              # Pass/Fail labels

# Step 2: Train Logistic Regression model
model = LogisticRegression()
model.fit(X, y)

# Step 3: Predict for new inputs
X_test = np.array([[2.5], [3.5], [5.5],[7.5]])
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)

print("Predicted Class (Pass=1/Fail=0):", y_pred)
print("Predicted Probabilities:\n", y_prob)

# Step 4: Visualize Sigmoid Curve
X_line = np.linspace(0, 7, 100).reshape(-1, 1)
y_line = model.predict_proba(X_line)[:, 1]  # Probability of class 1

plt.scatter(X, y, color='blue', label='Training Data')
plt.plot(X_line, y_line, color='red', label='Sigmoid Curve')
plt.axhline(0.5, color='green', linestyle='--', label='Decision Boundary (0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```
